```python
import pandas as pd
import numpy as np
```

```python
daily_data_file_path = 'daily_data.csv'
sleep_data_file_path = 'sleep_data.csv'

daily_df = pd.read_csv(daily_data_file_path)
sleep_df = pd.read_csv(sleep_data_file_path)
```

Perform merge operation


```python
new_df = pd.DataFrame.merge(daily_df, sleep_df, how='left', on='date')
```

```python
new_df.head()
```

Check len of columns

```python
print(len(daily_df))
print(len(new_df))
```

Check if data is the same

```python
sleep_df[sleep_df['date'] == '29-Jul-2019']
```

```python
new_df[new_df['date'] == '29-Jul-2019']
```

Move column positions

```python
# Column names
# 'sleep_hrs','sleep_mins','deep_sleep_mins','light_sleep_mins','rem_sleep_mins','awake_mins','unmeasurable_sleep_mins'

sleep_hrs = new_df.pop('sleep_hrs')
new_df.insert(1, 'sleep_hrs', sleep_hrs, False)

sleep_mins = new_df.pop('sleep_mins')
new_df.insert(2, 'sleep_mins', sleep_mins, False)

deep_sleep_mins = new_df.pop('deep_sleep_mins')
new_df.insert(3, 'deep_sleep_mins', deep_sleep_mins, False)

light_sleep_mins = new_df.pop('light_sleep_mins')
new_df.insert(4, 'light_sleep_mins', light_sleep_mins, False)

rem_sleep_mins = new_df.pop('rem_sleep_mins')
new_df.insert(5, 'rem_sleep_mins', rem_sleep_mins, False)

awake_mins = new_df.pop('awake_mins')
new_df.insert(6, 'awake_mins', awake_mins, False)

unmeasurable_sleep_mins = new_df.pop('unmeasurable_sleep_mins')
new_df.insert(7, 'unmeasurable_sleep_mins', unmeasurable_sleep_mins, False)
```
